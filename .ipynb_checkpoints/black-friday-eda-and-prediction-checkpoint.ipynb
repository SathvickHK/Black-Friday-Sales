### Finding and Treating Outliers

```python
out_cols=['Occupation','Product_Category_1','Purchase']
```

```python
def find_outliers(data):
    q1=data.quantile(0.25)
    q3=data.quantile(0.75)
    IQR=q3-q1
    LL = q1 - (1.5 * IQR)
    UL = q3 + (1.5 * IQR)
    #outliers = data[((data<(q1-1.5*IQR)) | (data>(q3+1.5*IQR)))]
    ul_count=df[data>UL].shape[0]
    ll_count=df[data<LL].shape[0]
    outlier_count = ul_count + ll_count
    outlier_perc=(outlier_count/len(df))*100
    return outlier_perc
```

```python
outlier_list=[]
res=0
for col in out_cols:
    val=df[col]
    res=find_outliers(val)
    outlier_list.append([col,res])
print(outlier_list)
```

```python
# Remving the outlier for purchase column
q1=df['Purchase'].quantile(0.25)
q3=df['Purchase'].quantile(0.75)
iqr=q3-q1
UL=q3+1.5*iqr
LL=q1-1.5*iqr
df.loc[df['Purchase']>UL,col]=UL
df.loc[df['Purchase']<LL,col]=LL
```

```python
plt.figure(figsize=(5,4))
sns.boxplot(data=df,x=df['Purchase'])
```

```python

df['Gender']=df['Gender'].map({'M':0,'F':1})

city_cat=pd.get_dummies(df['City_Category'],drop_first=True)
city_cat.rename(columns={'B':'City_B','C':'City_C'},inplace=True)
df=pd.concat([df,city_cat],axis=1)

dummy_col=pd.get_dummies(df['Age'],drop_first=True)
dummy_col.rename(columns={'18-25':'Age_18-25','26-35':'Age_26-35','36-45':'Age_36-45','46-50':'Age_46-50','51-55':'Age_51-55','55+':'Age_55+'},inplace=True)
df=pd.concat([df,dummy_col],axis=1)

df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].replace('4+','4')
```

```python
df.drop(['City_Category','Age'],axis=1,inplace=True)
```

```python
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].astype(int)
```

```python
df['Product_ID']=df['Product_ID'].str.replace('P00','').astype(int)
```

### Data scaling

```python
#Spliting the data for train and test

x=df.drop(['Purchase'],axis=1)
y=df['Purchase']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=42)
```

### Scaling the data for train and test data

```python
ss=StandardScaler()

x_train_cp=x_train.copy()
x_test_cp=x_test.copy()

print(df.columns)

num_cols=['Gender', 'Product_ID','Occupation', 'Stay_In_Current_City_Years','Marital_Status', 'Product_Category_1', 'Product_Category_2',
       'Product_Category_3','City_B', 'City_C', 'Age_18-25', 'Age_26-35', 'Age_36-45','Age_46-50', 'Age_51-55', 'Age_55+']

x_train[num_cols]=ss.fit_transform(x_train[num_cols])
x_test[num_cols]=ss.transform(x_test[num_cols])
```

```python
x_train.head(10)
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()

lr.fit(x_train,y_train)
```

```python
print(lr.intercept_)
print(lr.coef_)
```

```python
y_predict_lr=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
r2_scr=r2_score(y_test,y_predict_lr)
adj_r2_lr=1-(1-r2_scr)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
mse_lr=mean_squared_error(y_test,y_predict_lr)
```

```python
print(r2_scr)
print(adj_r2_lr)
print(mse_lr)
```

### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
from sklearn.model_selection import RandomizedSearchCV
```

```python
rf=RandomForestRegressor(random_state=42)
params={
    'max_depth':np.arange(3,10),
    'n_estimators':np.arange(50,500,100),
    'max_features': ['auto', 'sqrt','log2']
}

rand_cv=RandomizedSearchCV(rf,param_distributions=params,n_iter=5,scoring='r2',cv=10,n_jobs=3)
rand_cv.fit(x_train,y_train)
best_parm=rand_cv.best_params_
best_scr=rand_cv.best_score_
```

```python
print(best_parm)
print(best_scr)

rf.set_params(max_depth=6,n_estimators=50,max_features='auto')
```

```python
rf.fit(x_train,y_train)
y_pred_rf=rf.predict(x_test)
r2_rf=r2_score(y_test,y_pred_rf)
adj_r2=1-(1-r2_score(y_test, y_pred_rf))*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
mse_rf=mean_squared_error(y_test,y_pred_rf)
```

```python
print(r2_rf)
print(adj_r2)
print(mse_rf)
```

### XGBOOST

```python
from xgboost import XGBRFRegressor
xgb=XGBRFRegressor()

params={
    'max_depth':[3,4,5,6,7,8],
    'n_estimators':np.arange(50,500,100),
    'learning_rate':[.05,.1,.2,.5,.8,.9],
    'subsample': np.linspace(.3,.9,10),
    'min_child_weight':[2,3,4,5],
    'colsample_bytree':[.3,.4,.5,.6,.7,.8,.9],
    'gamma':[.2,.3,.4,.5,.6,.7,.8,.9]
}
```

```python
rncv_xgb=RandomizedSearchCV(xgb,param_distributions=params,cv=5,n_iter=5,n_jobs=3)
rncv_xgb.fit(x_train,y_train)
pram_list=rncv_xgb.best_params_
best_esti=rncv_xgb.best_estimator_
```

```python
print(pram_list)
print(best_esti)

xgb_reg=XGBRFRegressor(max_depth=7,n_estimators=250,learning_rate=.9,subsample=.56,min_child_weight=3,colsample_bytree=.8,gamma=.7)
```

```python
xgb_reg.fit(x_train,y_train)
```

```python
y_pred_xg=xgb_reg.predict(x_test)
r2_xg=r2_score(y_test,y_pred_xg)
adj_r2=1-(1-r2_score(y_test, y_pred_xg))*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
mse_rf=mean_squared_error(y_test,y_pred_xg)
```

```python
print(r2_xg)
print(adj_r2)
print(mse_rf)
```
